```python
goruped=df.groupby("Year")
coral_yearly_mean=df.groupby('Year')[top_coral].mean()
print(coral_yearly_mean)
```

Lets visualize these top corals and how they have grown or gone down




```python


plt.figure(figsize=(12, 6))

for coral in coral_yearly_mean.columns:
    plt.plot(coral_yearly_mean.index, coral_yearly_mean[coral], marker='o', label=coral)

plt.title('Mean Percent Cover of Top Coral Species Over Time')
plt.xlabel('Year')
plt.ylabel('Mean Percent Cover (%)')
plt.legend(title='Coral Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Despite being one of the top five corals in terms of mean percent cover and overall net increase, Orbicella annularis complex experienced a consistent decline from 2010 to 2020. This suggests a more complex temporal pattern, where the species may have suffered losses due to environmental stressors during the decade, followed by a limited recovery in recent years.
Siderastrea siderea exhibited a consistent upward trend in percent cover from 2010 to 2020, indicating its resilience and potential competitive advantage in the face of environmental changes. Its growth trajectory contrasts with the decline of more sensitive species such as Orbicella annularis complex, suggesting a possible shift in coral community structure.

Now let us see the how corals have changed with respect to stations over the years 

```python
stony_coral_cols = [
    'Acropora_cervicornis', 'Acropora_palmata', 'Agaricia_fragilis', 'Agaricia_lamarcki',
    'Cladocora_arbuscula', 'Colpophyllia_natans', 'Dendrogyra_cylindrus', 'Dichocoenia_stokesii',
    'Diploria_labyrinthiformis', 'Eusmilia_fastigiata', 'Favia_fragum', 'Helioseris_cucullata',
    'Isophyllia_rigida', 'Isophyllia_sinuosa', 'Madracis_aurentenra', 'Madracis_decactis_complex',
    'Manicina_areolata', 'Meandrina_meandrites', 'Millepora_alcicornis', 'Millepora_complanata',
    'Montastraea_cavernosa', 'Mussa_angulosa', 'Mycetophyllia_aliciae', 'Mycetophyllia_ferox',
    'Mycetophyllia_lamarckiana_complex', 'Oculina_diffusa', 'Oculina_robusta',
    'Orbicella_annularis_complex', 'Phyllangia_americana', 'Porites_astreoides',
    'Porites_porites_complex', 'Pseudodiploria_clivosa', 'Pseudodiploria_strigosa',
    'Scleractinia', 'Scolymia_sp', 'Siderastrea_radians', 'Siderastrea_siderea',
    'Solenastrea_bournoni', 'Stephanocoenia_intersepta', 'Undaria_agaricites_complex'
]
station_yearly_cover=df.groupby(['Year','StationID'])[stony_coral_cols].sum()
station_yearly_cover['Total_Stony_Coral_Cover'] = station_yearly_cover.sum(axis=1)
print(station_yearly_cover)
```

```python
cover_by_station=station_yearly_cover['Total_Stony_Coral_Cover'].reset_index()
#print(cover_by_station)
cover_pivot=cover_by_station.pivot(index='Year',columns='StationID',values='Total_Stony_Coral_Cover')
#print(cover_by_year)
top_stations = cover_pivot.mean().sort_values(ascending=False).head(5).index.tolist()
print(top_stations)
```

```python
cover_pivot[top_stations].plot(marker='o', figsize=(12, 6))
plt.title("the coral coverage over the years in stations ")
#here we are taking only top stations as there are many stations 
plt.xlabel('year')
plt.ylabel('coverage')
plt.grid(True)
plt.legend(title='StationID')
plt.tight_layout()
plt.show()
```
