```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
species_counts_columns = ['Acropora_cervicornis', 'Acropora_palmata', 'Acropora_prolifera',
                          'Agaricia_fragilis', 'Agaricia_lamarcki', 'Cladocor_arbuscula',
                          'Colpophyllia_natans', 'Dendrogyra_cylindrus', 'Dichocoenia_stokesii',
                          'Diploria_labyrinthiformis', 'Eusmilia_fastigiata', 'Favia_fragum',
                          'Helioseris_cucullata', 'Isophyllia_rigida', 'Isophyllia_sinuosa',
                          'Madracis_aurentenra', 'Madracis_decactis', 'Madracis_pharensis',
                          'Madracis_scenaria', 'Madracis_sp', 'Manicina_areolata',
                          'Meandrina_meandrites', 'Millepora_complanata', 'Montastraea_cavernosa',
                          'Mussa_angulosa', 'Mycetophyllia_aliciae', 'Mycetophyllia_danaana',
                          'Mycetophyllia_ferox', 'Mycetophyllia_lamarckiana', 'Mycetophyllia_sp',
                          'Oculina_diffusa', 'Oculina_robusta', 'Orbicella_annularis_complex',
                          'Orbicella_faveolata', 'Orbicella_franksi', 'Phyllangia_americana',
                          'Porites_astreoides', 'Porites_branneri', 'Porites_porites',
                          'Pseudodiploria_clivosa', 'Pseudodiploria_strigosa', 'Scleractinia',
                          'Scolymia_cubensis', 'Scolymia_lacera', 'Siderastrea_radians',
                          'Siderastrea_siderea', 'Solenastrea_bournoni', 'Solenastrea_hyades',
                          'Stephanocoenia_intersepta', 'Undaria_agaricites_complex']

species_richness = counts_df.groupby('StationID')[species_counts_columns].apply(lambda x: (x > 0).sum(axis=1).iloc[0]).reset_index(name='species_richness')
print("Species Richness per Station:")
print(species_richness.head())
print(density_df.columns)
```

```python
species_density_columns = ['Acropora_cervicornis', 'Acropora_palmata', 'Acropora_prolifera',
                           'Agaricia_fragilis', 'Agaricia_lamarcki', 'Cladocora_arbuscula',
                           'Colpophyllia_natans', 'Dendrogyra_cylindrus', 'Dichocoenia_stokesii',
                           'Diploria_labyrinthiformis', 'Eusmilia_fastigiata', 'Favia_fragum',
                           'Helioseris_cucullata', 'Isophyllia_rigida', 'Isophyllia_sinuosa',
                           'Madracis_aurentenra', 'Madracis_decactis', 'Madracis_pharensis',
                           'Madracis_scenaria', 'Madracis_sp', 'Manicina_areolata',
                           'Meandrina_meandrites', 'Millepora_complanata', 'Montastraea_cavernosa',
                           'Mussa_angulosa', 'Mycetophyllia_aliciae', 'Mycetophyllia_danaana',
                           'Mycetophyllia_ferox', 'Mycetophyllia_lamarckiana', 'Mycetophyllia_sp',
                           'Oculina_diffusa', 'Oculina_robusta', 'Orbicella_annularis_complex',
                           'Orbicella_faveolata', 'Orbicella_franksi', 'Phyllangia_americana',
                           'Porites_astreoides', 'Porites_branneri', 'Porites_porites',
                           'Pseudodiploria_clivosa', 'Pseudodiploria_strigosa', 'Scleractinia',
                           'Scolymia_cubensis', 'Scolymia_lacera', 'Siderastrea_radians',
                           'Siderastrea_siderea', 'Solenastrea_bournoni', 'Solenastrea_hyades',
                           'Stephanocoenia_intersepta', 'Undaria_agaricites_complex']



# Select only numeric columns from the species density columns
numeric_density_cols = density_df[species_density_columns].select_dtypes(include=['float64', 'int64']).columns.tolist()

# Calculate total stony coral density per StationID using only numeric columns
if numeric_density_cols:
    total_density = density_df.groupby('StationID')[numeric_density_cols].apply(lambda x: x.sum(axis=1)).reset_index(name='total_stony_coral_density')
    print("Total Stony Coral Density per Station:")
    print(total_density.head())
else:
    print("No numeric density columns found to sum.")
    total_density = pd.DataFrame() # Create an empty DataFrame if no numeric columns

# Proceed with merging and plotting only if total_density was calculated
if not total_density.empty:
    # Merge richness and density DataFrames
    richness_density_merged = pd.merge(species_richness, total_density, on='StationID', how='inner')

    # Create the scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(richness_density_merged['total_stony_coral_density'], richness_density_merged['species_richness'])
    plt.xlabel('Total Stony Coral Density')
    plt.ylabel('Stony Coral Species Richness')
    plt.title('Relationship between Stony Coral Density and Species Richness')
    plt.grid(True)
    plt.show()

    # Calculate the correlation coefficient
    correlation = richness_density_merged['total_stony_coral_density'].corr(richness_density_merged['species_richness'])
    print(f"Correlation between density and richness: {correlation:.2f}")
else:
    print("Skipping merging and plotting due to missing density data.")
```

Calculated Species Richness: Using the CREMP_SCOR_Summaries_2023_Counts.csv dataset, we grouped the data by each monitoring station (StationID). For each station, we counted the number of different stony coral species that had a count greater than zero. This gave us a measure of species richness at each station.

Calculated Total Stony Coral Density: Using the CREMP_SCOR_Summaries_2023_Density.csv dataset, we grouped the data by StationID again. For each station, we summed the density values of all the individual stony coral species present. This provided us with a measure of the total stony coral density at each station.

Merged the Data: We then combined the species richness and total density information into a single table using StationID as the common link.

Visualized the Relationship: We created a scatter plot with total stony coral density on one axis and species richness on the other. Each point on the plot represented a single monitoring station. This allowed us to visually inspect the relationship between these two variables.

Quantified the Relationship: We calculated the Pearson correlation coefficient, which gave us a numerical measure of the linear relationship between density and richness. A positive correlation suggests that higher density tends to be associated with higher richness, and the strength of the correlation indicates how strong this linear trend is.

In essence, we explored whether sites with more stony coral (higher density) also tend to have a greater variety of stony coral species (higher richness) in the 2023 data. The scatter plot and the correlation coefficient helped us understand the nature and strength of this relationship.
